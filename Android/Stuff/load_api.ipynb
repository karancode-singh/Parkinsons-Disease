```python
def load_model(X, Y, Z, Timestamp):
    x_test = forge_make(X, Y, Z, Timestamp)
    loaded_model = pickle.load(open(filename, 'rb'))
    #print(x_test)
    y_pred = loaded_model.predict_proba(x_test)
    return y_pred
```
